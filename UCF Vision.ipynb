```python
def tprint(mat):
    for row in mat:
        print(row)
```

```python
img = cv2.imread("/home/vinaykudari/Documents/code/practice/vision/sample-images/logo.jpg", 1)
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_grey, cmap='gray')
plt.show()
```

```python
#somoothning and gradient
sobel_x = [[1, 0, -1], [2, 0, -2], [1, 0, -1]]
sobel_y = np.array(sobel_x).transpose()
tprint(sobel_y)
```

```python
img_box_x = convolve2d(img_grey, np.array(sobel_x), mode='same')
plt.imshow(img_box_x, cmap='gray')
plt.show()
```

```python
img_box_y = convolve2d(img_grey, np.array(sobel_y), mode='same')
plt.imshow(img_box_y, cmap='gray')
plt.show()
```

```python
for i in range(img_box_x.shape[0]):
    for j in range(img_box_x.shape[1]):
        img_box[i][j] = math.sqrt(img_box_x[i][j]**2 + img_box_y[i][j]**2)
```
